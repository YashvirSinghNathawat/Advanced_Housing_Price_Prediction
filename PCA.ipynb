```python
# Select only numeric columns


# Standardize the features
x = StandardScaler().fit_transform(result)

# Apply PCA
pca = PCA(n_components=None)  # Set n_components to None to keep all components
principal_components = pca.fit_transform(x)

# Determine the number of components that explain at least 95% of the variance
explained_variance = pca.explained_variance_ratio_
cumulative_variance = explained_variance.cumsum()
num_components = (cumulative_variance >= 0.97).argmax() + 1

print(f"Number of principal components needed to explain at least 95% variance: {num_components}")

# Apply PCA with the determined number of components
pca = PCA(n_components=num_components)
x_reduced = pca.fit_transform(x)

# Create a DataFrame with the reduced principal components
principal_df_reduced = pd.DataFrame(data=x_reduced, columns=[f'PC{i+1}' for i in range(x_reduced.shape[1])])

# Print the reduced DataFrame
print("\nDataFrame with principal components explaining at least 98% of the variance:")
print(principal_df_reduced)
```
